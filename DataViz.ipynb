```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.boxplot(data=df, x='Department', y='MonthlyIncome')

# Adding labels and title
plt.xlabel('Department')
plt.ylabel('Monthly Income')
plt.title('Monthly Income Distribution by Department')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
g = sns.FacetGrid(df, col="Department", col_wrap=4, hue = "Department", palette='Set2', height=4)

# Map a histogram plot onto the grid
g.map(plt.hist, 'MonthlyIncome', bins=20)  # Adjust the number of bins as needed

# Adding labels and title
g.set_axis_labels('Monthly Income', 'Frequency')
g.fig.suptitle('Histograms of Monthly Income by Department')

# Adjust the layout
plt.tight_layout()
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.histplot(df['Age'], bins=20, kde=False, color  = "yellow")

mean_value = df['Age'].mean()
plt.axvline(x=mean_value, color='r', linestyle='--', label=f'Mean: {mean_value:.2f}')

# Adding labels and title
plt.xlabel('Total Working Years')
plt.ylabel('Frequency')
plt.title('Histogram of Total Working Years')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.boxplot(data=df, x='Gender', y='Age', palette='Set2')

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxplot of Age by Gender')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.boxplot(data=df, x='Gender', y='Age', palette='Set2')


# Calculate mean age for each gender
mean= df['Age'].mean()


plt.axhline(mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxplot of Age by Gender')

# Show the plot
plt.show()
```
